2) Write a python program to display IMDB’s Top rated 100 movies’ data (i.e. name, rating, year 
of release) and make data frame.

```python
page=requests.get("https://www.imdb.com/list/ls055592025/")
page
```

```python
soup=BeautifulSoup(page.content)
soup
```

```python
first_title=soup.find('h3',class_='lister-item-header')
first_title.text
```

```python
first_title.text.replace('\n','')
```

```python
Rating=soup.find('div',class_='ipl-rating-star small')
Rating.text
```

```python
Rating.text.replace('\n','')
```

```python
movies=[]

for i in soup.find_all('h3',class_="lister-item-header"):
    movies.append(i.text.replace('\n',''))

movies
```

```python
Rating=[]

for i in soup.find_all('div',class_='ipl-rating-star small'):
    Rating.append(i.text.replace('\n',''))
```

```python
Rating
```

```python
import pandas as pd

IMDB_100=pd.DataFrame({'Top 100 movies':movies,'Rating':Rating})
```

```python
IMDB_100
```

Write python program to display list of respected former finance minister of India

```python
page=requests.get("https://presidentofindia.nic.in/former-presidents")
page
```

```python
soup=BeautifulSoup(page.content)
soup
```

```python
President=soup.find('div',class_='desc-sec')
President
```

```python
President.text
```

```python
President.text.replace('\n','')
```

```python
President=[]

for i in soup.find_all('div',class_='desc-sec'):
    President.append(i.text.replace('\n',''))

President
```

```python
import pandas as pd

President=pd.DataFrame({'President':President})
President
```

```python
page=requests.get("https://www.dineout.co.in/hyderabad-restaurants/biryani-lover")
page
```

```python
Restaurant=soup.find('div',class_="restnt-info cursor")
Restaurant
```

```python
Cuisine=soup.find('div',class_='detail-info')
Cuisine
```

```python
Location=soup.find('div',class_='restnt-loc ellipsis')
Location
```

```python
Location.text
```

```python
Ratings=soup.find('div',class_="restnt-rating rating-4")
Ratings
```

```python
Ratings.text
```

```python
Image_url=soup.find('img',class_='no-img')
Image_url
```

```python
Restaurant=[]

for i in soup.find_all('div',class_='restnt-info cursor'):
    Restaurant.append(i.text)
    
Restaurant
```

```python
Cuisine=[]

for i in soup.find_all('div',class_='detail-info'):
    Cuisine.append(i.text)
    
Cuisine
```

```python
Location=[]

for i in soup.find_all('div',class_='restnt-loc ellipsis'):
    Location.append(i.text)
    
Location
```

```python
Rating=[]

for i in soup.find_all('div',class_='restnt-rating rating-4'):
    Rating.append(i.text)
    
    
Rating
```

```python
Image=[]

for i in soup.find_all('img',class_='no-img'):
    Image.append(i['data-src'])
    
    
Image
```

```python
import pandas as pd

Restaurant_Dineout=pd.DataFrame({'Restaurant':Restaurant,'Cuisine':Cuisine,'Location':Location,'Rating':Rating,'Image':Image})
Restaurant_Dineout
```
