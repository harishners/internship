# Seperating features and label

```python
x=df.drop('W',axis=1)
y=df['W']
```

# Standard Scalarization

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
x=pd.DataFrame(scaler.fit_transform(x),columns=x.columns)
x
```

# Variance Inflation Factor

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor
```

```python
vif=pd.DataFrame()
vif['VIF values']=[variance_inflation_factor(x.values,i) for i in range(len(x.columns))]
vif['Features']=x.columns
vif
```

As ER is having high VIF value, we'll drop it.

```python
x.drop('ER',axis=1,inplace=True)
```

```python
vif=pd.DataFrame()
vif['VIF values']=[variance_inflation_factor(x.values,i)
                  for i in range(len(x.columns))]
vif['Features']=x.columns
vif
```

let's drop RA as it has high VIF values

```python
x.drop('RA',axis=1,inplace=True)
```

```python
vif=pd.DataFrame()
vif['VIF values']=[variance_inflation_factor(x.values,i)
                  for i in range(len(x.columns))]
vif['Features']=x.columns
vif
```

We've fixed the multi colinearilty issue by using Variance Inflation factor. Now let's proceed with model building

# Model Building

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score, mean_squared_error, mean_absolute_error
```

```python
maxAccu=0
maxRs=0

for i in range(1,200):
    x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.30,random_state=i)
    LR=LinearRegression()
    LR.fit(x_train,y_train)
    pred=LR.predict(x_test)
    acc=r2_score(y_test,pred)
    if acc>maxAccu:
        maxAccu=acc
        maxRs=i
print('Best R2_Score is',maxAccu,'at random state',maxRs)
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.30,random_state=maxRs)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import GradientBoostingRegressor, AdaBoostRegressor, ExtraTreesRegressor
from sklearn.tree import DecisionTreeRegressor
from sklearn.linear_model import Lasso, Ridge
from sklearn.svm import SVR
```

```python
LR=LinearRegression()
LR.fit(x_train,y_train)
pred=LR.predict(x_test)
predtrain=LR.predict(x_train)
print('R2 score at test data:',r2_score(y_test,pred))
print('R2 score at train data:',r2_score(y_train,predtrain)*100)
print('mean squared error:',mean_squared_error(y_test,pred))
print('mean absolute error:',mean_absolute_error(y_test,pred))
print('root mean squared error:',np.sqrt(mean_squared_error(y_test,pred)))
```

```python
plt.figure(figsize=(10,6))
plt.scatter(x=pred,y=y_test,color='r')
plt.plot(pred,pred,color='b')
plt.xlabel('Actual',fontsize=15)
plt.ylabel('Predicted',fontsize=15)
plt.title('Linear Regression',fontsize=15)
plt.show()
```

```python
GBR=GradientBoostingRegressor()
GBR.fit(x_train,y_train)
pred_GBR=GBR.predict(x_test)
predtrain_GBR=GBR.predict(x_train)
print('R2 score at test data:',r2_score(y_test,pred_GBR))
print('R2 score at train data:',r2_score(y_train,predtrain_GBR)*100)
print('mean squared error:',mean_squared_error(y_test,pred_GBR))
print('mean absolute error:',mean_absolute_error(y_test,pred_GBR))
print('root mean squared error:',np.sqrt(mean_squared_error(y_test,pred_GBR)))
```

```python
ABR=AdaBoostRegressor()
ABR.fit(x_train,y_train)
pred_ABR=ABR.predict(x_test)
predtrain_ABR=ABR.predict(x_train)
print('R2 score at test data:',r2_score(y_test,pred_ABR))
print('R2 score at train data:',r2_score(y_train,predtrain_ABR)*100)
print('mean squared error:',mean_squared_error(y_test,pred_ABR))
print('mean absolute error:',mean_absolute_error(y_test,pred_ABR))
print('root mean squared error:',np.sqrt(mean_squared_error(y_test,pred_ABR)))
```

```python
ETR=ExtraTreesRegressor()
ETR.fit(x_train,y_train)
pred_ETR=ETR.predict(x_test)
predtrain_ETR=ETR.predict(x_train)
print('R2 score at test data:',r2_score(y_test,pred_ETR))
print('R2 score at train data:',r2_score(y_train,predtrain_ETR)*100)
print('mean squared error:',mean_squared_error(y_test,pred_ETR))
print('mean absolute error:',mean_absolute_error(y_test,pred_ETR))
print('root mean squared error:',np.sqrt(mean_squared_error(y_test,pred_ETR)))
```

```python
ls=Lasso()
ls.fit(x_train,y_train)
pred_LS=ls.predict(x_test)
predtrain_LS=ls.predict(x_train)
print('R2 score at test data:',r2_score(y_test,pred_LS))
print('R2 score at train data:',r2_score(y_train,predtrain_LS)*100)
print('mean squared error:',mean_squared_error(y_test,pred_LS))
print('mean absolute error:',mean_absolute_error(y_test,pred_LS))
print('root mean squared error:',np.sqrt(mean_squared_error(y_test,pred_LS)))
```

```python
rd=Ridge()
rd.fit(x_train,y_train)
pred_Rd=rd.predict(x_test)
predtrain_Rd=rd.predict(x_train)
print('R2 score at test data:',r2_score(y_test,pred_Rd))
print('R2 score at train data:',r2_score(y_train,predtrain_Rd)*100)
print('mean squared error:',mean_squared_error(y_test,pred_Rd))
print('mean absolute error:',mean_absolute_error(y_test,pred_Rd))
print('root mean squared error:',np.sqrt(mean_squared_error(y_test,pred_Rd)))
```

```python
DTR=DecisionTreeRegressor()
DTR.fit(x_train,y_train)
pred_DTR=DTR.predict(x_test)
predtrain_DTR=DTR.predict(x_train)
print('R2 score at test data:',r2_score(y_test,pred_DTR))
print('R2 score at train data:',r2_score(y_train,predtrain_DTR)*100)
print('mean squared error:',mean_squared_error(y_test,pred_DTR))
print('mean absolute error:',mean_absolute_error(y_test,pred_DTR))
print('root mean squared error:',np.sqrt(mean_squared_error(y_test,pred_DTR)))
```

```python
svr=SVR()
svr.fit(x_train,y_train)
pred_svr=svr.predict(x_test)
predtrain_svr=svr.predict(x_train)
print('R2 score at test data:',r2_score(y_test,pred_svr))
print('R2 score at train data:',r2_score(y_train,predtrain_svr)*100)
print('mean squared error:',mean_squared_error(y_test,pred_svr))
print('mean absolute error:',mean_absolute_error(y_test,pred_svr))
print('root mean squared error:',np.sqrt(mean_squared_error(y_test,pred_svr)))
```

```python
from sklearn.neighbors import KNeighborsRegressor as KNN

knn=KNN()
knn.fit(x_train,y_train)
pred_knn=knn.predict(x_test)
predtrain_knn=knn.predict(x_train)
print('R2 score at test data:',r2_score(y_test,pred_knn))
print('R2 score at train data:',r2_score(y_train,predtrain_knn)*100)
print('mean squared error:',mean_squared_error(y_test,pred_knn))
print('mean absolute error:',mean_absolute_error(y_test,pred_knn))
print('root mean squared error:',np.sqrt(mean_squared_error(y_test,pred_knn)))
```

From the above models, it looks Linear Regression is the best model but let's proceed further to make sure which is the best model.
